```
# 爬蟲範例列表 https://github.com/FinMind/FinMind/tree/master/Data 

import datetime
import requests
import pandas as pd
url = 'http://finmindapi.servebeer.com/api/data'
list_url = 'http://finmindapi.servebeer.com/api/datalist'
translate_url = 'http://finmindapi.servebeer.com/api/translation'

from pandas_datareader import data as web
import fix_yahoo_finance as yf
yf.pdr_override()
```

```

def Get_US_Stock_Price_Data(stock_id , Dates = 20 ):
  '''取得歷史股價, 如果Dates沒有輸入, 則只取20天內的數據'''
  StartDate = datetime.datetime.now() - datetime.timedelta(days=Dates) 
  form_data = {'dataset':'USStockPrice','stock_id':stock_id,'date':StartDate}
  res = requests.post(
      url,verify = True,
      data = form_data)
  temp = res.json()
  data = pd.DataFrame(temp['data'])
  data = data.set_index('date')
  return data

#單元測試: 
test_data = Get_US_Stock_Price_Data('TSLA')
test_data
```

```
def Get_US_ETF_Price_Data(stock_id , Dates = 20 ):
  '''取得歷史ETF價格, 如果Dates沒有輸入, 則只取20天內的數據'''
  start = datetime.datetime.now() - datetime.timedelta(days=Dates-1)
  end = datetime.datetime.now()- datetime.timedelta(days=1)
  ETFdata = web.get_data_yahoo([stock_id],start, end)
  ETFdata['stock_id'] = 'VOO'
  return ETFdata

#單元測試: 
_data = Get_US_ETF_Price_Data('VOO')
_data
```

```
# 選股列表
STOCK_ID_LIST = ['TSLA', 'AAPL','AMAT','CSCO','DIS','INTC', 'MSFT','QCOM']
ETF_ID_LIST = ['VOO'] 

Days = 300

# 合併價格資料, 以利後續計算處裡
merge_date = pd.DataFrame() 
for i, each_stock_ID in enumerate( STOCK_ID_LIST) :
  data = Get_US_Stock_Price_Data(each_stock_ID , Days)  
  if i == 0 :
    merge_date[each_stock_ID] = data['Close']
  else:
    merge_date[each_stock_ID] = data['Close']
print(merge_date)

# 合併ETF資料 
for i, each_ETF_ID in enumerate(ETF_ID_LIST):
    ETFdata = Get_US_ETF_Price_Data(each_ETF_ID , Days)   
    merge_date[each_ETF_ID] = ETFdata['Close']
 
merge_date['CASH'] = 1 
print(merge_date.head(4))
```

資產配置:

```
import pandas as pd
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns

# Read in price data
#df = pd.read_csv("tests/stock_prices.csv", parse_dates=True, index_col="date")
df = merge_date 
# Calculate expected returns and sample covariance
mu = expected_returns.mean_historical_return(df)
S = risk_models.sample_cov(df)

# Optimise for maximal Sharpe ratio
ef = EfficientFrontier(mu, S )

weights = ef.max_sharpe()
ef.portfolio_performance(verbose=True)

print(  round(pd.Series(weights),4) *100 )
print( mu )
```

```
# 不同風險模型

S = risk_models.CovarianceShrinkage(df)
ef = EfficientFrontier(mu, S.ledoit_wolf() ,  gamma=1)

weights = ef.max_sharpe()
ef.portfolio_performance(verbose=True)

print(  round(pd.Series(weights),4) *100  )
print( mu )
```
